```python
#Exemple 4.11

import numpy as np
import matplotlib.pyplot as plt

# Definir os valores de X1
X1 = np.arange(0, 1.1, 0.1)

# Definir constantes
T = 75  # Temperatura em Celsius

# Presso de vapor do acetonitrila (P1) e nitrometano (P2)
P1 = np.exp(14.2724 - 2945.47 / (T + 224))  # acetonitrila
P2 = np.exp(14.2043 - 2972.64 / (T + 209))  # nitrometano

# Definir valores iniciais
Pt = (0.1 / 0.01) / 120  # Presso total arbitrria para m = 1/11

# Inicializar vetores para armazenar resultados
Press = []
Y1 = []

# Loop para calcular os valores
for x in X1:
    K1 = P1 / Pt  # Coeficiente de acetonitrila
    K2 = P2 / Pt  # Coeficiente de nitrometano
    
    soma = K1 * x + K2 * (1 - x)  # Calcular a soma
    
    if soma < 1:
        break  # Se a soma for menor que 1, interromper o loop
    
    press_atual = Pt * soma  # Calcular a presso
    Press.append(press_atual)  # Armazenar a presso
    y1_atual = K1 * x / soma  # Calcular Y1
    Y1.append(y1_atual)  # Armazenar Y1

# Plotar os resultados
plt.plot(X1[:len(Press)], Press, 'k-+', label='Presso vs X1')
plt.plot(X1[:len(Press)], Y1, 'k-*', label='Y1 vs X1')
plt.axis([0, 1, 20, 100])  # Definir limites dos eixos

# Adicionar rtulos e ttulo
plt.xlabel('x,y para acetonitrila')
plt.ylabel('Pt (kPa)')
plt.title('Diagrama Pxy do sistema acetonitrila/nitrometano')
plt.legend()

# Mostrar o grfico
plt.show()
```
