```python
import numpy as np

#Definindo a matriz

a11, a12, a13 = 0.2, -0.5, -0.2 #0.2F1 + (-0.5)F3 + (-0.2)F4 = -250 [BALANÇO DE A]
a21, a22, a23 = 0.3, -0.3, -0.4 #0.3F1 + (-0.3)F3 + (-0.4)F4 = -250 [BALANÇO DE B]
a31, a32, a33 = 0.5, -0.2, -0.4 #0.5F1 + (-0.2)F3 + (-0.4)F4 = 0    [BALANÇO DE D]

# Definindo a matriz dos coeficientes 
A = np.array([
    [a11, a12, a13],  
    [a21, a22, a23],  
    [a31, a32, a33]   
])

# Definindo o vetor dos termos constantes

b1,b2,b3 = -250, -250, 0

# Definindo a matriz de solução
S = np.array([[b1],
              [b2],
              [b3]])

# Calculando o determinante de A
det_A = (a11 * (a22 * a33 - a23 * a32) -
         a12 * (a21 * a33 - a23 * a31) +
         a13 * (a21 * a32 - a22 * a31))
det_A
# Calculando a matriz dos cofatores
C = np.array([[a22*a33 - a23*a32, -(a21*a33 - a23*a31), a21*a32 - a22*a31],
              [-(a12*a33 - a13*a32), a11*a33 - a13*a31, -(a11*a32 - a12*a31)],
              [a12*a23 - a13*a22, -(a11*a23 - a13*a21), a11*a22 - a12*a21]])

# Transpondo a matriz dos cofatores para obter a matriz adjunta
adj_A = C.T

# Calculando a inversa de A
A_inv = adj_A / det_A

#Multiplicando a matriz inversa pela matriz de saída
X = np.dot(A_inv, S)

print("A solução do sistema é:", X)
det_A
```
