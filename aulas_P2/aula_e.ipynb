#### Criação de Matrizes Sparsas em Python

A maioria dos problemas de Engenharia que apresentam várias entradas e várias saídas em que a resolução é feita aplicando métodos matriciais caem em soluções que possuem muitos zeros armazenados. Resoluções computacional de matrizes com muitos zeros causam excessivo desgaste computacional para no final o resultado ser zero. Dessa forma, podemos dizer que: se a maioria dos elementos de uma matriz tem $0$ é chamado de **matriz esparsa**. Os dois principais benefíios do uso de matriz esparsa em vez de uma matriz simples são:
- **Armazenamento**: Se existem menos elementos diferentes de zero armazenados nas células, então os cálculos serão realizados apenas com esses elementos. Dessa forma, menos memória pode ser usada para armazenar esses elementos;
- **Tempo de Computação**: O tempo de computação pode ser economizado projetando-se lógicamente uma estrutura de dados que atravesse apenas elementos diferentes de zero.

As matrizes esparsas são, geralmente, utilizadas em aprendizado de máquina aplicada, como em dados contendo codificações de dados que mapeiam categorias para contar e também em subcampos inteiros de aprendizado de máquina, como **Processamento de Linguagem Natural** (NL).

Resolver uma matriz esparsa por uma matriz $2D$ leva ao desperdício de muita memória, pois os **zeros na matriz são inúteis** na maioria dos casos. Portanto, em vez de armazenar zeros com elementos diferentes de zero, aramazenamos apenas elementos diferentes de zero. Isso significa armazenar elementos diferentes de zero com triplos (**linhas**, **colunas**, **valor**).


#### Biblioteca SciPy para computação

O **ScyPy** do Python fornece ferramentas para criação de matrizes esparsas usando várias estruturas de dados, bem como ferramentas para converter uma matriz densa em uma matriz esparsa. A função **csr_matrix()** é usada para criação de uma matriz **esparsa no formato de linhas esparsas compactadas**, enquanto **csc_matrix()** é usada para a criação de uma **matriz esparsa no formato de colunas esparsas compactadas.**

#### Notações:

**csr_matrix()**: cria uma matriz esparsa em linhas (**row** = linhas)
<br>**csc_matrix()**: cria uma matriz esparsa em colunas (**col** = colunas)

#### Sintax:

scipy.sparse.csr_matrix(**shape-None**, **dtype-None**)

#### Parâmetros:
**shape**: obtem a forma de uma matriz
<br>**dtype**: tipo de dados da matriz

#### Criação de uma matriz de **Coo**rdenadas - **COO**
**Fonte**: https://docs.scipy.org/doc/scipy/reference/sparse.html


Uma das matrizes mais intitivas é a matriz esparsa de coordenadas, podemos criar uma matriz esparsa de COO rapidamente, **basicamente precisamos das coordenadas de elementos diferentes de zero na matriz esparsa**. Para criar uma **coo_matrix**, precisamos de 3 matrizes NumPy unidimensionais. O primerio *array* representa os índices de linhas, o segundo *array* representa os índices da coluna e o terceiro *array* represena os dados diferents de zero no elemento. Os índices de linhas e colunas especificam a localização do elemento diferente de zero e a matriz de dados especifica os dados reais diferentes de zero.

Vamos criar uma matriz esparsa no formato de COO usando um exemplo simples, conforme será apresentado no exercício a seguir.

2 - Vamos criar os dados que correspondem aos valores armazenados nas células com valores diferentes de zero.


```python
display(Image(filename='tabela_basica_2.png', width=600))
```

```python
import numpy as np
from scipy import sparse
from scipy.stats import uniform

# criando os índices de linhas (row):
row_ind = np.array([0, 1, 1, 3, 4])

# criando os índices da coluna (columns)
col_ind = np.array([0, 2, 4, 3, 4])

# criando os índices dos dados (data):
data = np.array([1, 2, 3, 4, 5], dtype=float)
```

```python
# Agora vamos criar a matriz esparsa COO dadas as três informações acima.
mat_coo = sparse.coo_matrix((data, (row_ind, col_ind)))

print(mat_coo)
```
